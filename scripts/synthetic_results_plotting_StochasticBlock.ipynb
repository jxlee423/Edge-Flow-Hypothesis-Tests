# 2. Stochastic Block

## 2.1 Different AB Ratio (Stochastic Block, Cov_Sign = Positive)

```python
specific_result_folder = "Stochasticblock_AB_Pos"
plot_name = "figure7_power_abratio_stochasticblock_pos"

full_results_path = os.path.join(results_to_analyze_dir_SB, specific_result_folder)
filename_base = os.path.join(output_plots_dir, plot_name)

run_analysis_and_generate_plots({
        "root_dir": full_results_path,
        "group_by_col": "A",
        "legend_title": "A:B Ratio",
        "palette": {5: '#377eb8', 10: '#ff7f00', 15: '#4daf4a', 20: '#e41a1c', 25: '#984ea3'},
        "main_title": 'Power vs. Number of Nodes for Different A:B Ratios with 95% CI (Cov_Sign = Pos)',
        "caption": ("Figure 7: This line chart illustrates the relationship between the number of nodes and the statistical power "
                    "of the hypothesis test, categorized by different A:B Ratios."
                    "\n*Flows are all generated by the same setting: Distance = 5, Cov_sign = Pos, Rho = 0.5, Nc = 3, Graph = StochasticBlock.\n"
                    "#Classification: Top 25% Edge Degree to class1; Bottom 25% Edge Degree to class2."),
        "filename_base": filename_base
    })
```

Looking at the Overall (Corrected) graph, the lines for parameters 10-25 are clustered together and are plotted with a much higher power than the line with a ratio of 5, which starts at lower power and increases at a slower rate. I interpreted this as increasing the ratio from 5 to 10 drastically changes the number of inter- vs intra-community edges. With a ratio of 5, inter vs intra-community edge density may be similar. But increasing the ratio to 10 results in a graph where the number of intra-community edges overpowers the number of inter-community edges, more clearly resembling the stochastic block structure where the number of intra-community edges is dense and the number of inter-community edges is sparse. As a result of this clearer stochastic structure, power quickly increases.
