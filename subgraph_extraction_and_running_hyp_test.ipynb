```python
def convert_graph_format(input_file, output_file):
    with open(input_file, 'r') as infile, open(output_file, 'w') as outfile:
        for line in infile:
            # Strip whitespace and split the line by comma
            parts = line.strip().split(',')
            # Join the parts with a tab and write to the output file
            outfile.write('\t'.join(parts) + '\n')

def process_folder(input_folder, output_folder):
    # Ensure output folder exists
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # List all files in the input folder
    for filename in os.listdir(input_folder):
        if filename.endswith('.txt'):  # Check if the file is a .txt file
            input_file = os.path.join(input_folder, filename)
            output_file = os.path.join(output_folder, f"{os.path.splitext(filename)[0]}_converted.txt")

            # Check if the file is a file and not a directory
            if os.path.isfile(input_file):
                convert_graph_format(input_file, output_file)


# Example usage:
input_folder = source_path
output_folder = output_path

process_folder(input_folder, output_folder)
```

```python
import os

def process_file(input_filename, output_filename):
    with open(input_filename, 'r') as file, open(output_filename, 'w') as output_file:
        for line in file:
            # Split on the first colon and take everything after it
            if ':' in line:
                node_ids = line.split(':', 1)[1].strip()
                nodes = node_ids.split()
                if len(nodes) >= 4:  # Ensure at least four nodes for the expected relationships
                    # Extract required nodes
                    first_node = nodes[0]
                    second_node = nodes[1]
                    third_node = nodes[2]
                    fourth_node = nodes[3]
                    # Format the output with the additional relationships
                    output_line = (f"{node_ids}: {first_node} - {second_node}, "
                                   f"{first_node} - {third_node}, "
                                   f"{first_node} - {fourth_node}, "
                                   f"{second_node} - {third_node}, "
                                   f"{second_node} - {fourth_node}, "
                                   f"{third_node} - {fourth_node}\n")
                else:
                    output_line = node_ids + "\n"  # If there aren't enough nodes, just output the node ids
                output_file.write(output_line)

def process_directory(input_directory, output_directory):
    # Ensure the output directory exists
    if not os.path.exists(output_directory):
        os.makedirs(output_directory)

    for filename in os.listdir(input_directory):
        if filename.endswith("_filtered_sub.txt"):
            input_path = os.path.join(input_directory, filename)
            output_filename = filename.replace("_filtered_sub.txt", "_edge_lists.txt")
            output_path = os.path.join(output_directory, output_filename)
            process_file(input_path, output_path)

# Directories for input and output
input_directory = '/content/drive/MyDrive/Colab_Notebooks/gtrieScanner_src_01/test_output/Erdos/size_4_output'
output_directory = '/content/drive/MyDrive/Colab_Notebooks/gtrieScanner_src_01/test_output/Erdos/size_4_edge_lists'

# Process all files in the directory
process_directory(input_directory, output_directory)
```

```python
import shutil
import os

def add_row_ids_to_file(file_path, temp_file_path):
    """Append a row ID to each line in the temporary file."""
    updated_lines = []
    with open(file_path, 'r') as file:
        for line_num, line in enumerate(file, start=1):
            updated_line = f"{line.strip()}\t{line_num}\n"
            updated_lines.append(updated_line)

    with open(temp_file_path, 'w') as file:
        file.writelines(updated_lines)

def load_edge_mapping(file_path):
    """Load the mapping of node pairs to row IDs (edge IDs)."""
    mapping = {}
    with open(file_path, 'r') as file:
        for line in file:
            parts = line.strip().split('\t')
            if len(parts) >= 3:  # Ensure there are at least three parts (edge_id, node1, node2)
                edge_id = parts[-1].strip()  # Row ID is the last part
                nodes = parts[:-1]  # All but the last are considered nodes
                if len(nodes) == 2:  # We expect exactly two nodes
                    node1 = nodes[0].strip()
                    node2 = nodes[1].strip()
                    node_pair = tuple(sorted([node1, node2]))
                    mapping[node_pair] = edge_id
    return mapping

def find_edge_id(node1, node2, mapping):
    """Find the edge ID (row ID) for the given pair of nodes."""
    node_pair = tuple(sorted([node1.strip(), node2.strip()]))
    return mapping.get(node_pair, "NaN")

def process_edge_lists(edge_list_path, mapping, output_path):
    """Process the edge list file and append edge IDs based on the mapping."""
    with open(edge_list_path, 'r') as infile, open(output_path, 'w') as outfile:
        for line in infile:
            parts = line.split(':')
            if len(parts) > 1:
                prefix = parts[0]
                edge_list = parts[1].strip()
                edges = edge_list.split(', ')
                new_edges = []
                for edge in edges:
                    node1, node2 = edge.split('-')
                    edge_id = find_edge_id(node1, node2, mapping)
                    new_edges.append(f"({node1} - {node2}, {edge_id})")
                new_line = f"{prefix}: {', '.join(new_edges)}\n"
                outfile.write(new_line)

# Base paths
converted_files_base_path = '/content/drive/MyDrive/Colab_Notebooks/gtrieScanner_src_01/test_output/Erdos/converted_files'
edge_lists_base_path = '/content/drive/MyDrive/Colab_Notebooks/gtrieScanner_src_01/test_output/Erdos/size_4_edge_lists'
flow_lists_base_path = '/content/drive/MyDrive/Colab_Notebooks/gtrieScanner_src_01/test_output/Erdos/flow_lists'

# Ensure the output directory exists
os.makedirs(flow_lists_base_path, exist_ok=True)

# Iterate over each row and column
for row in range(1, 21):
    for col in range(1, 5):
        converted_file_path = f"{converted_files_base_path}/matrix_row{row}_col{col}_converted.txt"
        edge_list_file_path = f"{edge_lists_base_path}/matrix_row{row}_col{col}_converted_edge_lists.txt"
        output_file_path = f"{flow_lists_base_path}/matrix_row{row}_col{col}_edges.txt"
        temp_file_path = f"{converted_files_base_path}/matrix_row{row}_col{col}_converted_temp.txt"

        # Step 1: Create a temporary file with row IDs added, without modifying the original file
        shutil.copyfile(converted_file_path, temp_file_path)
        add_row_ids_to_file(converted_file_path, temp_file_path)

        # Step 2: Load mappings from the temporary file with the new row IDs
        edge_mapping = load_edge_mapping(temp_file_path)

        # Step 3: Process the edge list file with the updated mappings
        process_edge_lists(edge_list_file_path, edge_mapping, output_file_path)

        # Cleanup: Remove the temporary file
        os.remove(temp_file_path)
```

```python
import os
import re

def process_subgraph(subgraph):
    try:
        nodes_str, edges_str = subgraph.split(": ", 1)
    except ValueError:
        return ""

    nodes = list(map(int, nodes_str.split()))
    edge_pattern = re.compile(r"\(\s*(\d+)\s*-\s*(\d+)\s*,\s*(\d+|NaN)\s*\)")
    valid_edges = []
    node_counts = {}

    for match in edge_pattern.finditer(edges_str):
        node1, node2, edge_id = match.groups()
        if edge_id != "NaN":
            node1, node2, edge_id = int(node1), int(node2), int(edge_id)
            valid_edges.append((node1, node2, edge_id))
            for node in (node1, node2):
                node_counts[node] = node_counts.get(node, 0) + 1

    if not valid_edges:
        return ""

    sorted_nodes = sorted(node_counts.keys(), key=lambda x: (node_counts[x], x))
    node_order = {node: index for index, node in enumerate(sorted_nodes)}

    def edge_sort_key(edge):
        n1, n2 = edge[:2]
        return (node_order[max(n1, n2)], node_order[min(n1, n2)], n1, n2)

    valid_edges.sort(key=edge_sort_key)
    edge_ids = [str(eid) for _, _, eid in valid_edges]
    return " ".join(edge_ids)

def process_file(input_path):
    with open(input_path, 'r') as file:
        lines = file.readlines()

    processed_lines = [process_subgraph(line.strip()) for line in lines if line.strip()]

    # Write directly back to the same input file, overwriting original content
    with open(input_path, 'w') as file:
        file.write("\n".join(line for line in processed_lines if line))

# Directory containing the files to process
input_directory = '/content/drive/MyDrive/Colab_Notebooks/gtrieScanner_src_01/test_output/Erdos/flow_lists'

# Process each file in the directory
for filename in os.listdir(input_directory):
    if filename.endswith("_edges.txt"):  # Process only files that match the pattern
        input_file_path = os.path.join(input_directory, filename)
        process_file(input_file_path)  # Process and overwrite the file directly
```

```python
import csv
import numpy as np
import os
import pandas as pd

# Base paths
flow_lists_path = '/content/drive/MyDrive/Colab_Notebooks/gtrieScanner_src_01/test_output/Erdos/flow_lists'
flows_path = '/content/drive/MyDrive/Colab_Notebooks/gtrieScanner_src_01/test_output/Erdos/flows'

# Process each file in the flow_lists directory
for filename in os.listdir(flow_lists_path):
    if filename.endswith("_edges.txt"):  # Process only sorted .txt files
        flowsss_path = os.path.join(flow_lists_path, filename)

        # Correcting filename parsing
        row_col_part = filename.split('_edges.txt')[0]  # e.g., matrix_row20_col4
        row_part = row_col_part.split('_col')[0].split('_row')[-1]  # e.g., '20' from 'matrix_row20'
        col_part = row_col_part.split('_col')[-1]  # e.g., '4' from 'matrix_row20_col4'

        csv_filename = f"matrix_{row_part}_{col_part}.xlsx"  # Correctly formatting the filename
        csv_path = os.path.join(flows_path, csv_filename)
        output_path = os.path.join(flow_lists_path, f"matrix_{row_part}_{col_part}_processed_flows.csv")

        # Step 1: Read the edges from .txt file
        with open(flowsss_path, 'r') as file:
            edges_list = [list(map(int, line.strip().split())) for line in file.readlines()]

        # Step 2: Load the flow data from the corresponding .xlsx file
        if not os.path.exists(csv_path):
            continue  # Skip processing if file not found

        flow_data = pd.read_excel(csv_path, header=None).to_numpy()

        # Step 3: Map each edge ID in .txt to the corresponding flow values in .xlsx
        processed_flows = []
        for edges in edges_list:
            row_data = []
            for column in range(flow_data.shape[1]):  # Iterate over each column
                edge_flows = []
                for edge_id in edges:
                    if 0 <= edge_id - 1 < flow_data.shape[0]:
                        edge_flows.append(flow_data[edge_id - 1][column])
                    else:
                        edge_flows.append(np.nan)  # Use NaN if no match found
                row_data.extend(edge_flows)
                row_data.append('')  # Add empty column as separator
            processed_flows.append(row_data[:-1])  # Remove the last empty column

        # Step 4: Save the processed flows to a new .csv file
        with open(output_path, 'w', newline='') as file:
            writer = csv.writer(file)
            writer.writerows(processed_flows)

        # Step 5: Delete the original .txt file after processing
        os.remove(flowsss_path)
```

```python
import pandas as pd
import numpy as np
import os
import bootstrapped.bootstrap as bs
import bootstrapped.stats_functions as bs_stats

# Base paths
data_directory = '/content/drive/MyDrive/Colab_Notebooks/gtrieScanner_src_01/test_output/Erdos/flow_lists'
output_directory = '/content/drive/MyDrive/Colab_Notebooks/gtrieScanner_src_01/test_output/Erdos/CI'
os.makedirs(output_directory, exist_ok=True)  # Ensure output directory exists

# Number of bootstrap samples and significance level
bootstrap_samples = 200
significance_level = 0.05

def bootstrap_covariance(data, num_iterations=200):
    """Bootstrap the entire covariance matrix calculation."""
    n = len(data)
    bootstrapped_matrices = [np.cov(data[np.random.choice(n, n, replace=True)], rowvar=False) for _ in range(num_iterations)]
    # Calculating the percentile for lower and upper bounds of the CI
    lower_bounds = np.percentile(bootstrapped_matrices, 100 * (significance_level / 2), axis=0)
    upper_bounds = np.percentile(bootstrapped_matrices, 100 * (1 - significance_level / 2), axis=0)
    return lower_bounds, upper_bounds

# Process each CSV file in the data directory
for filename in os.listdir(data_directory):
    if filename.endswith('.csv'):
        data_path = os.path.join(data_directory, filename)
        output_file_path = os.path.join(output_directory, filename.replace('.csv', '_covariance_matrices_with_intervals.txt'))

        # Load the data from the CSV file
        try:
            data = pd.read_csv(data_path, header=None)
            if data.empty:
                continue  # Skip empty data files
        except pd.errors.EmptyDataError:
            continue  # Skip files that can't be loaded

        # Open the output file
        with open(output_file_path, 'w') as output_file:
            # Calculate covariance matrices for each group of 4 columns (assuming columns_per_group = 4)
            for i in range(0, data.shape[1], 5):  # Skip separator columns if present
                group_data = data.iloc[:, i:i+4].dropna().to_numpy()
                if group_data.shape[1] == 4:  # Ensure proper matrix dimensions
                    covariance_matrix = np.cov(group_data, rowvar=False)
                    if covariance_matrix.shape == (4, 4):  # Validate matrix size
                        output_file.write("Covariance Matrix:\n")
                        np.savetxt(output_file, covariance_matrix, fmt='%.6f')

                        # Generate bootstrap confidence intervals
                        lower_bounds, upper_bounds = bootstrap_covariance(group_data, bootstrap_samples)

                        # Write the confidence intervals to the file
                        output_file.write("Bootstrap Confidence Intervals for Covariance Matrix:\n")
                        for row in range(4):
                            for col in range(4):
                                output_file.write(f"Element [{row},{col}]: {lower_bounds[row, col]:.6f}, {upper_bounds[row, col]:.6f}\n")
                        output_file.write("\n")
```

```python
import numpy as np
import pandas as pd
import os
import re

# Directory where CI files are stored
ci_directory = '/content/drive/MyDrive/Colab_Notebooks/gtrieScanner_src_01/test_output/Erdos/CI'
results_directory = '/content/drive/MyDrive/Colab_Notebooks/gtrieScanner_src_01/test_output/Erdos/CI_Results'
os.makedirs(results_directory, exist_ok=True)  # Ensure results directory exists

# Process each file in the CI directory
for filename in os.listdir(ci_directory):
    if filename.endswith('.txt'):
        file_path = os.path.join(ci_directory, filename)
        result_file_path = os.path.join(results_directory, f'results_{filename}')

        with open(file_path, 'r') as file:
            content = file.readlines()

        matrices = []
        current_matrix = []

        for line in content:
            if 'Covariance Matrix' in line:
                if current_matrix:
                    matrices.append(current_matrix)
                current_matrix = []
            elif 'Element [' in line:
                match = re.search(r'Element \[(\d+),(\d+)\]: ([\d\.-]+), ([\d\.-]+)', line)
                if match:
                    row, col = int(match.group(1)), int(match.group(2))
                    lower, upper = float(match.group(3)), float(match.group(4))
                    current_matrix.append(((row, col), (lower, upper)))

        if current_matrix:
            matrices.append(current_matrix)

        with open(result_file_path, 'w') as result_file:
            for idx, matrix_cis in enumerate(matrices):
                # Extract CIs for diagonal entries
                diagonal_cis = [(lower, upper) for ((row, col), (lower, upper)) in matrix_cis if row == col]

                # Check diagonal overlaps by comparing max lower to min upper across diagonals
                if diagonal_cis:
                    max_lower = max(lower for (lower, upper) in diagonal_cis)
                    min_upper = min(upper for (lower, upper) in diagonal_cis)
                    diagonal_overlaps = max_lower <= min_upper
                else:
                    diagonal_overlaps = True  # Assume true if no diagonal entries

                # Other checks remain the same
                zero_in_zero_one = any(lower <= 0 <= upper for ((row, col), (lower, upper)) in matrix_cis if row == 0 and col == 1)
                rho_sigma_cis = [(lower, upper) for ((row, col), (lower, upper)) in matrix_cis if (row, col) in [(0, 2), (0, 3), (1, 2), (1, 3), (2, 3)]]
                rho_sigma_overlaps = True
                if rho_sigma_cis:
                    max_lower = max(lower for (lower, upper) in rho_sigma_cis)
                    min_upper = min(upper for (lower, upper) in rho_sigma_cis)
                    rho_sigma_overlaps = max_lower <= min_upper

                if diagonal_overlaps and zero_in_zero_one and rho_sigma_overlaps:
                    result_file.write(f"Not reject for matrix {idx + 1} in file {filename}\n")
                else:
                    rejection_reasons = []
                    if not diagonal_overlaps:
                        rejection_reasons.append("diagonal case rejected")
                    if not zero_in_zero_one:
                        rejection_reasons.append("zero case for (0,1) rejected")
                    if not rho_sigma_overlaps:
                        rejection_reasons.append("rho*sigma^2 case rejected")
                    result_file.write(f"REJECT for matrix {idx + 1} ({', '.join(rejection_reasons)}) in file {filename}\n")
```
